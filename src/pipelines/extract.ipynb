```python
import pandas as pd
import yfinance as yf

tickers = ['AAPL', 'MSFT']
historical_data = []
year = 2023
for ticker in tickers:
    stock = yf.Ticker(ticker)
    info = stock.info
    start_date = f"{year}-11-30"
    end_date = f"{year}-12-31"
    try:
        hist = stock.history(start=start_date, end=end_date)
        for index, row in hist.iterrows():
            data = {
                "company_id": info.get("uuid", ""),
                "date": index.date(),
                "ticker": ticker,
                "price": row['Close'],
                "industry": info.get("industry", "")
            }
            historical_data.append(data)
    except Exception as e:
        print(f"Error extracting data for {ticker}: {e}")
        continue

df = pd.DataFrame(historical_data)
df.head(5)
```

```python
company = yf.Ticker(ticker)

income_statement = company.financials.T  
balance_sheet = company.balance_sheet.T  

combined_df = pd.concat([income_statement, balance_sheet], axis=1)
combined_data = pd.DataFrame({
            'year': combined_df.index,
            'ticker': ticker,
            'total_revenue': combined_df['Total Revenue'],
            'gross_profit': combined_df['Gross Profit'],
            'net_income': combined_df['Net Income'],
            'total_debt': combined_df['Total Debt']
        })
combined_data.head(5)


# income_statement = income_statement.reindex(index=income_statement.index.union(balance_sheet.index))
# balance_sheet = balance_sheet.reindex(index=income_statement.index)

# print(income_statement.head())
# print(balance_sheet.head())
```
